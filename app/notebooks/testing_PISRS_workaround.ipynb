```python
# Import necessary libraries
import asyncio
import nest_asyncio
from pyppeteer import launch

# Apply nest_asyncio
nest_asyncio.apply()


website_url = "http://www.pisrs.si/pregledPredpisa?id=ZAKO1603"
button_html_signature = 'button[data-test="result-oblike-pdf"]'


# Define the async function
async def main():
    browser = await launch(headless=True)
    page = await browser.newPage()
    
    request_urls = []

    def handle_request(request):
        url = request.url
        print(request)
        print(url)
        if 'PDF' in url:  # Assuming PDF in the URL indicates a relevant request
            request_urls.append(url)
            print("Captured PDF URL:", url)
    
    page.on('request', handle_request)

    await page.goto(website_url)
    await page.click(button_html_signature)
    await asyncio.sleep(5)  # Wait to ensure all requests are captured
    
    await browser.close()
    return request_urls

# Execute the function
urls = await main()
print("All Captured URLs:", urls)
```
